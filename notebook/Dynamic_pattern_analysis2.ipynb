### K-means on compressed features

We create a new sparse matrix for our components. Each component is now represented by a sparse input vector of input_ids and the probability to have an activated node for each component

```python
cluster_count = 12
```

```python
B = create_compressed_feature_matrix(filtered_components).toarray()
print 'Compressed feature matrix:', B.shape
clusters = None

# normalize data by norm along each row
b_norms = np.apply_along_axis(np.linalg.norm, 1, B)
B = B.astype(float) / b_norms[:, np.newaxis]
```

```python
clusters = None
A = create_binary_feature_matrix(filtered_components)
```

```python
path = os.path.join(NOTEBOOK_DIR, 'km_cfeat_' + str(cluster_count) + '.pickle')
def cluster_data(feat_matrix, k):
    km = sk.cluster.KMeans(n_clusters=k, init='k-means++',
                           n_init=50, max_iter=300, tol=0.0001,
                           precompute_distances=True, verbose=0, random_state=None, copy_x=True, n_jobs=4)
    return km.fit_predict(B)
```

```python
clusters = cluster_data(B, cluster_count)
```

Save clusters on disk

```python
# Save features
with open(path, 'w') as f:
    pickle.dump(clusters, f)
```
